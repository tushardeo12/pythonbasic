SCALING THE DATA

```python
from sklearn.preprocessing import MinMaxScaler  ##importing MinMaxScaler
```

```python
scaler=MinMaxScaler()  ## initializing the MinMaxScaler() function
```

```python
scaler.fit(data)  ## fitting the data in the intitilaized function
```

here the process happening is the scaling of the data and hence the entire data set is passed in the fit() fucntion.
a single column cant be passed here in the fit() function becoz the data scaling is a process that is applied to the entire data . it cant be applied on seleceted columns.

hence the entire data that has been preprocessed before feeding to the MinMaxScaler() must be passed throgh it so that the data scaling can be applied to the entire data set

```python
scaled_data=pd.DataFrame(data=scaler.transform(data),columns=data.columns,index=data.index)

## here the columns and the index is passed because the netire data needs to be named in the printing part, hence the
## entire column and indexhave been passed inside the function 
```
