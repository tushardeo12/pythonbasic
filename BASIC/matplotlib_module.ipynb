MATPLOTLIB(used for plotting graphs)

```python
## inporting modules required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#  %matplotlib inline  

##The inline option with the %matplotlib magic function renders the plot out cell even
##if show() function of plot object is not called.
```

LINE PLOT

```python
x=[2,4,5,7,8]  ## taking two list of values or coordinates
y=[3,12,4,12,5]


plt.plot(x,y)  ## plot() plts the graph
plt.show  ## shows the plotted graph
```

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]
plt.plot(x,y,"^")  ## this prints the coordinates marking "^" at the coordinate position
plt.plot(x,y)  ## thi prints line throgh the coordinates
plt.show     ## shows the plotted graph
```

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]

plt.title("line graph")  ## this function is used to give the graph a name or tittle

plt.xlabel("x")  ## to name the x axis
plt.ylabel("y")  ## to name the y axis

plt.plot(x,y) ## plots the graph by passing the x and y value throuh the plot function
```

SCATTER PLOT GRAPH

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]

plt.scatter(x,y) ## .scatter() prints the scatter plot graph
```

COLOR, LINEWIDTH AND ANNOTATIONS

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]  ## first set of coordinates

x2=[2,5,8,5,12]
y2=[7,8,4,6,9]  ## second set of coordinates

plt.annotate(xy=[2,4],s="blue") ## annotate() marsk the line with a name or label. [2,4] marks the coordinates of the label
plt.annotate(xy=[2,7],s="red")

# plt.legend(["red","blue"],loc=4)  ## prints the legend for the graph

plt.plot(x,y,color="b",linewidth=3)   ## color sets the color of the line and linewidth sets the width
plt.plot(x2,y2,color="r",linewidth=3)  
plt.show    ## shows the plotted graph
```

 BAR GRAPHS  (VERTICAL)

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]
plt.bar(x,y) ## .bar() plots the bar graph for the given coordinates vvertically
plt.show
```

BAR GRAPH HORIZONTAL

```python
x=[2,4,5,7,8]
y=[3,12,4,12,5]
plt.barh(x,y)  ##  .barh() plots the bar graph for the given coordinates horizontly
plt.show
```

PIE CHART

```python
x=[2,4,5,7,8]  ## taking values

plt.pie(x)  ## .pie() plots the pie chart graph 
plt.show   


```

```python
x=[2,4,5,7,8]  ## taking values

lb=["a","b","c","d","e"]   ## this lines defines the labesl for the pie chart ,, lb is a variable

b=[0,0.1,0,0,0]  ## this line defines the explosion for the pie chart ,, b is sa variable

plt.pie(x,labels=lb,explode=b)    ## labels and explode are passed throgh the .pie()  function
## labels marks the labels for the different areas in the pie chart
## explode moves any piece of the pie cgart outside the chart


plt.show
```

```python
x=[2,4,5,7,8]  ## taking values

lb=["a","b","c","d","e"]   ## this lines defines the labesl for the pie chart ,, lb is a variable

b=[0,0.1,0,0,0]  ## this line defines the explosion for the pie chart ,, b is sa variable

plt.pie(x,labels=lb,explode=b,shadow=True)    ## labels and explode are passed throgh the .pie()  function
## labels marks the labels for the different areas in the pie chart
## explode moves any piece of the pie cgart outside the chart
## shadow adds a bit of shadow to the entire pie chart

plt.show
```
