WAYS OF ENCODING

        LABEL ENCODING -it replaces the object with a numerical vlaue 

```python
from sklearn.preprocessing import LabelEncoder  ## importing required module
```

```python
Label_encoder=LabelEncoder()   ## initializing the LabelEncoder() function
```

```python
Label_encoder_data=Label_encoder.fit_transform(data["Sex"])  ## fitting the required data using fit_transform() function
```

in the cell above and below the fit_transform function is used
in the above cell a particular column ["sex"] is passed in the fit_transform() because here we are performing the label encoding of the data which means we are changing the data from object value to a numerical value and hence this kind of operation is possible to be performed on a single column .

because we might only need to convert a single column to the numerical value hence pasing the requires column in the function is posible.
but this cant be done for other things like the MinMaxScaler() used for data scaling becoz the data scaling is a type of operation that must be applied on the entire data set ,it cant be applid to only one particular chosen column.
hence in the case of data scaling  it is needed to pass the entire dataset in the fit_ transform function.
passing only a single column will give error

```python
Label_encoder_data=pd.DataFrame(data=Label_encoder_data,columns=["Sex"])  ## creating a dataframe from the encoded data
## here the columns=["sex"] is passed becoz if not done the data frame would show 0 instead of sex written at the top
```

in the above cell only ["sex"] column is passed in the function data frame becoz only that particular columns needs to be named in the printing area
